```python
# last year
initial_date = "2022-01-01T00:00:00.000Z"
final_date = "2023-05-01T23:59:59.999Z"

dfs = {}
for symbol in symbols[:10]:
    r = requests.get(f"http://DESKTOP-3V2V0BR.local:8000/prices?symbol={symbol}&timeframe=TIMEFRAME_M15&initial_date={initial_date}&final_date={final_date}")
    
    json_data = r.json()
    df = pd.DataFrame(json_data)
    df['datetime']= pd.to_datetime(df['datetime'])
    df = df.set_index("timestamp")
    
    dfs[symbol] = df
```

```python
dfs
```

```python
def slope_rolling(arr: np.ndarray):
    return arr[-1] - arr[0]

def get_slope(serie: pd.Series, period: int = 10):
    return serie.rolling(period, min_periods=period).apply(slope_rolling, raw=True)
```

```python
AAPL34 = dfs['AAPL34']
AAPL34['slope'] = get_slope(serie=AAPL34['close'], period=2)
```

```python
AAPL34["close1"] = (AAPL34.slope.shift(3) <= 0) & (AAPL34.slope.shift(2) > 0)
AAPL34.head(5)
```
