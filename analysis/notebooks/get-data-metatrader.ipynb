# Test Getting Data from Metatrader

```python
import requests
```

```python
r = requests.get('http://DESKTOP-3V2V0BR.local:8000/symbols?type=VISTA')
r.status_code
```

```python
symbols = r.json()
symbols
```

```python
symbol = "PETR3"
initial_date = "2022-01-01T00:00:00.000Z"
final_date = "2023-05-01T23:59:59.999Z"
r = requests.get(f"http://DESKTOP-3V2V0BR.local:8000/prices?symbol={symbol}&timeframe=TIMEFRAME_M15&initial_date={initial_date}&final_date={final_date}")
r.status_code
```

```python
json_data = r.json()
json_data
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(json_data)
df['datetime']= pd.to_datetime(df['datetime'])
df.set_index("timestamp")
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(df.datetime, df.close)
plt.xticks(rotation=90)
plt.ylabel('close')
plt.show()
```
